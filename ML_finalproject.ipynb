```python
import numpy as np

Q1 = np.percentile(df['current_value'], 25, axis=0)
Q3 = np.percentile(df['current_value'], 75, axis=0)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

dff = df[(df['current_value'] > lower_bound) & (df['current_value'] < upper_bound)]
```

```python
Q1 = np.percentile(dff['highest_value'], 25, axis=0)
Q3 = np.percentile(dff['highest_value'], 75, axis=0)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

dfff = dff[(dff['highest_value'] > lower_bound) & (dff['highest_value'] < upper_bound)]
```

```python
x=dfff.drop(['current_value'],axis=1)
y=dfff['current_value']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

scaler.fit(x_train)

x_train=scaler.transform(x_train)

x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
lin=LinearRegression()
rforest=RandomForestRegressor(criterion='poisson',n_estimators=100,random_state=0)
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error
lst=[lin,rforest]
```

```python
for i in lst:

  i.fit(x_train,y_train)

  y_pred=i.predict(x_test)

  print(i)

  print("r2: ",r2_score(y_test,y_pred))

  print('MAPE',mean_absolute_percentage_error(y_test,y_pred))
  print('*'*100)
```

```python
from sklearn.model_selection import GridSearchCV
para={'n_estimators':[50,100,150],'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],'random_state':[0,1,42]}
gcv=GridSearchCV(rforest,para,cv=10,scoring='neg_mean_absolute_percentage_error')
gcv.fit(x_train,y_train)
print(gcv.best_params_)
```
