```python
y=df.iloc[:,-1].values
y
```

```python
x=df.iloc[:,:-1].values      #.values make the df as arrays
x
```

```python
from sklearn.model_selection import train_test_split                      #spliting into testing data and training data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler     #normalisation
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
knn=KNeighborsClassifier(n_neighbors=7)
bays=GaussianNB()
svc=SVC()
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
lst=[knn,bays,svc]
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("MODEL IS.......",i)
  print(y_pred)
  print("*"*100)
  print("Confusion matrix is ...............")
  print(confusion_matrix(y_test,y_pred))
  print("Acuuracy score is ...............")
  print(accuracy_score(y_test,y_pred))
  print("Classification report is ...............")
  print(classification_report(y_test,y_pred))
  print("*"*100)
```
