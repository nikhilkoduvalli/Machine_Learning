```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
df
```

```python
df.isna().sum()
```

```python
df1=df.drop(['Id'],axis=1)
df1
```

```python
x=df1.iloc[:,:-1].values
x
```

```python
y=df1.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler   #normalisation
scaler=StandardScaler()                           #object for Standardscalar
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict(scaler.transform([[1,5.2,3.9,3]])))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python

score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
lab=['Iris-setosa', 'Iris-versicolor','Iris-virginica']
cmd=ConfusionMatrixDisplay(cm,display_labels=lab)
cmd.plot()
```
